```
def flatten_dict(dd, separator ='_', prefix =''): 
    return { prefix + separator + k if prefix else k : v 
             for kk, vv in dd.items() 
             for k, v in flatten_dict(vv, separator, kk).items() 
             } if isinstance(dd, dict) else { prefix : dd }
```

# Content Based Recommender System

```
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```
features= ['type','points','tags','rating']
```

```
def combine_features(row):
    return row['type']+' '+row['points']+' '+ row['tags']+' '+row['rating']
```

```
for feature in features:
    cf_prob_df[feature]= cf_prob_df[feature].fillna('')

cf_prob_df['combined_features']= cf_prob_df.apply(combine_features,axis=1)
```

```
print(cf_prob_df['combined_features'][0])
```

```
cv= CountVectorizer()
count_matrix= cv.fit_transform(cf_prob_df['combined_features'])
```

```
print(count_matrix)
```

```
cos_sim= cosine_similarity(count_matrix)
print(cos_sim)
```

```
def get_name_from_serialno(serial):
    return cf_prob_df[cf_prob_df.SerialNo==serial]['name'].values[0]

def get_serialno_from_name(name):
    return cf_prob_df[cf_prob_df.name==name]['SerialNo'].values[0]
```

```
prob_user_likes= 'Monster Invaders'
prob_serialno= get_serialno_from_name(prob_user_likes)
#print(prob_serialno)
similar_probs= list(enumerate(cos_sim[prob_serialno]))
```

```
sorted_similar_probs= sorted(similar_probs,key= lambda x:x[1], reverse=True)[1:]
```

```
i=0
print("Top 10 similar problems to "+prob_user_likes+" are:\n")
for element in sorted_similar_probs:
    print(get_name_from_serialno(element[0]))
    i=i+1
    if i>10:
        break
```
